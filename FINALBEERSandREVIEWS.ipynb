```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
beer = "beers.csv"
beersdf = pd.read_csv(beer)
beersdf.head()
```

```python
beersdf.dropna()
```

```python
beercols = ["id", "name", "brewery_id", "state", "country", "style", "availability", "abv"]
beersclean = beersdf[beercols].copy()
beersclean.set_index("id", inplace=True)
beersclean.head()
```

```python
cabeers = beersclean[beersclean["state"] == 'CA']
cabeers
```

```python
cabeers.to_csv("CaliforniaBeers.csv")
```

```python
reviews = "reviews.csv"
revdf = pd.read_csv(reviews)
revdf.head()
```

```python
revcols = ["beer_id", "date", "look", "smell", "taste", "feel", "overall", "score"]
revclean = revdf[revcols].copy()
revclean.columns = ["id", "date", "look", "smell", "taste", "feel", "overall", "score"]
revclean
```

```python
revcleaned = revclean.dropna()
revcleaned
```

```python
revc = revcleaned.groupby('id').mean()
revc
```

```python
cabeersandrev = cabeers.merge(revc, how="inner", on=["id"])
cabeersandrev
```

```python
del cabeersandrev["name"]
del cabeersandrev["brewery_id"]
del cabeersandrev["state"]
del cabeersandrev["country"]
del cabeersandrev["style"]
del cabeersandrev["availability"]
del cabeersandrev["abv"]
cabeersandrev
```

```python
CaliforniaReviews = cabeersandrev
CaliforniaReviews
```

```python
CaliforniaReviews.to_csv("CaliforniaReviews.csv")
```

```python
connection_string = "postgres:charlie@localhost:5432/Beers_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
cabeers.to_sql(name='beers', con=engine, if_exists='append', index=True)
```

```python
pd.read_sql_query('select * from beers', con=engine).head()
```

```python
CaliforniaReviews.to_sql(name='reviews', con=engine, if_exists='append', index=True)
```
