```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
breweries = "breweries.csv"
brewdf = pd.read_csv(breweries)
brewdf
```

```python
brewdf.dropna()
```

```python
brewcols = ["id", "name", "city", "state", "country", "types"]
brewclean = brewdf[brewcols].copy()
brewclean.set_index("id", inplace=True)
brewclean
```

```python
cabrews = brewclean[brewclean["state"] == 'CA']
cabrews
```

```python
cabrews.to_csv("CaliforniaBreweries.csv")
```

```python
connection_string = "postgres:charlie@localhost:5432/Beers_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
cabrews.to_sql(name='breweries', con=engine, if_exists='append', index=True)
```

```python
pd.read_sql_query('select * from breweries', con=engine).head()
```
