```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
brew = "breweries.csv"
brewdf = pd.read_csv(brew)
brewdf.head()
```

```python
beer = "beers.csv"
beersdf = pd.read_csv(beer)
beersdf.head()
```

```python
brewdf.dropna()
```

```python
beersdf.dropna()
```

```python
brewcols = ["id", "name", "city", "state", "country", "types"]
brewclean = brewdf[brewcols].copy()
brewclean.set_index("id", inplace=True)
brewclean.head()
```

```python
beercols = ["id", "name", "brewery_id", "state", "country", "style", "availability", "abv"]
beersclean = beersdf[beercols].copy()
beersclean.set_index("id", inplace=True)
beersclean.head()
```

```python
beersclean.dtypes
```

```python
connection_string = "postgres:charlie@localhost:5432/Beer"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
brewclean.to_sql(name='breweries', con=engine, if_exists='append', index=True)
```

```python
beersclean.to_sql(name='beers', con=engine, if_exists='append', index=True)
```

```python
pd.read_sql_query('select * from breweries', con=engine).head()
```

```python
pd.read_sql_query('select * from beers', con=engine).head()
```
