```python
import pandas as pd
```

```python
file = "reviews.csv"
```

```python
beer_df = pd.read_csv(file)
```

```python
beer_df.head()
```

```python
del beer_df['username']
del beer_df['text']
beer_df.head()
```

```python
beer_df.count()
```

```python
beer_df=beer_df.dropna(how='any')
beer_df.head()
```

```python
beer_df.count()
```

```python
beer_df.dtypes
```

```python
beer_df = beer_df.sort_values("beer_id", ascending=False)
beer_df.head()
```

```python
beer_index_df = beer_df.reset_index(drop=True)
beer_index_df.head(10)
```

```python
beer_index_df = beer_index_df.rename(columns={"beer_id":"Beer_ID", "date":"Date", "look":"Look", "smell":"Smell",\
                                             "taste":"Taste", "feel":"Feel", "overall":"Overall", "score":"Average_Score"})
beer_index_df.head(10)
```

```python
beer_index_df = pd.DataFrame(beer_index_df)
beer_index_df.head()
```

```python
beer_index_df.to_csv('final_beer_reviews_df.csv')
```
